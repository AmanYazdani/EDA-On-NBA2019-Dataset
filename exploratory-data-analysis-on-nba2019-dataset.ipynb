```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Web Scraping 

```python
url='https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
html= pd.read_html(url, header=0) #defining header=0 to indicate the first row is the header itself
html
```

```python
len(html)
```

the link contains only one table, in some cases there might be multiple tables we should choose from.

```python
df2019= html[0]
df2019
```

```python
df2019[df2019.Age=='Age']
```

It was observed on the website that the header is repeated every 20 rows. We searched for it here in the notebook and turns out we have 26 such headers in the data. We need to get rid of them as we only need one header.

```python
df= df2019.drop(df2019[df2019.Age=='Age'].index)
df
```

It can be observed that the shape of the data is reduced which means the extra headers were removed.

The null values are successfully taken care of.

**6.Which player had the highest 3-point field goals per game?**

```python
df[df['3P']==df['3P'].max()]
```

**7.Which player had the highest assists per game?**

```python
df[df.AST==df.AST.max()]
```

**8.Which player scored the highest PTS in the Los Angeles Lakers?**

```python
los=df.groupby('Tm').get_group('LAL')
los
```

Here, we created a new susbset lol that only contains data related to team los angeles lakers, now we can perform max func on the pts for this subset to get the answer.

```python
los[los['PTS']==los['PTS'].max()]
```

The answer is LeBron James.

**9.Of the 5 positions, which position scores the most points?**

```python
df.groupby('Pos').describe()
```

It can be seen that we have more than 5 positions because some players play style is hybrid we need to remove the hybrid positions and only consider the regular 5 positions.

```python
positions=['C','PF','SF','SG','PG']
POS=df[df['Pos'].isin(positions)]
POS
```

Doing this we were able to successfully remove the hybrid positions and created a new subset with the regualr 5 positions.

```python
POS.groupby('Pos')['PTS'].describe()
```
